```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import warnings


# Data Cleaning and Preprocessing
WRF_subdata.dropna(subset=['XLAT', 'XLONG'], inplace=True)

# Interpolate missing values
WRF_subdata.interpolate(method='linear', inplace=True)

# Extract row 182
data_phase_2 = pd.concat([pd.DataFrame(WRF_subdata.columns), WRF_subdata.iloc[181]], axis=1).T

# Export the data for manual handling
data_phase_2.to_csv('data_phase_2.csv', index=False)

# Load cleaned data
data_phase_2clean = pd.read_csv('Date_Phase2_Reorg.csv')
data_phase_2clean
```

```python
data_phase_2clean = data_phase_2clean.rename(columns={'Date/Time' : 'Date.Time'}) 
data_phase_2clean
```

```python
data_phase_2clean['Date.Time']
```

```python
data_phase_2clean['Date.Time'] = pd.to_datetime(data_phase_2clean['Date.Time'], format='%d.%m.%Y.%H.%M')

# Visualize data
plt.figure(figsize=(10, 6))
plt.plot(data_phase_2clean['Date.Time'], data_phase_2clean['TSK'], color='blue')
plt.title('Surface Temperature Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Surface Temperature')
plt.show()
```

```python


# Convert data to time series
WRF_ts = pd.Series(data_phase_2clean['TSK'].values, index=data_phase_2clean['Date.Time'])

# Check for stationarity
result = adfuller(WRF_ts)
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Split the time series into training and testing sets
n_train = int(len(WRF_ts) * 0.8)
train_WRF_ts = WRF_ts[:n_train]
test_WRF_ts = WRF_ts[n_train:]

# Fit ARIMA model
Arima_WRF_model = ARIMA(train_WRF_ts, order=(5, 1, 0))
Arima_WRF_model_fit = Arima_WRF_model.fit()
Arima_WRF_forecast = Arima_WRF_model_fit.forecast(len(test_WRF_ts))

# Evaluate ARIMA model
arima_rmse = np.sqrt(mean_squared_error(test_WRF_ts, Arima_WRF_forecast))
print('ARIMA Model RMSE:', arima_rmse)

# Visualize ARIMA forecast
plt.figure(figsize=(10, 6))
plt.plot(test_WRF_ts.index, test_WRF_ts.values, color='red', label='Actual')
plt.plot(test_WRF_ts.index, Arima_WRF_forecast, color='blue', label='ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Surface Temperature')
plt.title('ARIMA Forecast vs Actual')
plt.legend()
plt.show()
```

```python

# Linear Regression model
Linear_WRF_model = make_pipeline(StandardScaler(), LinearRegression())
Linear_WRF_model.fit(train_WRF_ts.index.to_numpy().reshape(-1, 1), train_WRF_ts)
linear_WRF_forecast = Linear_WRF_model.predict(test_WRF_ts.index.to_numpy().reshape(-1, 1))

# Evaluate Linear Regression model
linear_rmse = np.sqrt(mean_squared_error(test_WRF_ts, linear_WRF_forecast))
print('Linear Regression Model RMSE:', linear_rmse)

# Random Forest model
RandomForest_WRF_model = RandomForestRegressor()
RandomForest_WRF_model.fit(train_WRF_ts.index.to_numpy().reshape(-1, 1), train_WRF_ts)
RandomForest_WRF_forecast = RandomForest_WRF_model.predict(test_WRF_ts.index.to_numpy().reshape(-1, 1))

# Evaluate Random Forest model
rf_rmse = np.sqrt(mean_squared_error(test_WRF_ts, RandomForest_WRF_forecast))
print('Random Forest Model RMSE:', rf_rmse)

# Support Vector Regression models
svr_kernels = ['rbf', 'linear', 'poly']
svr_models = {}

for kernel in svr_kernels:
    svr_model = SVR(kernel=kernel)
    svr_model.fit(train_WRF_ts.index.to_numpy().reshape(-1, 1), train_WRF_ts)
    svr_models[kernel] = svr_model
    svr_forecast = svr_model.predict(test_WRF_ts.index.to_numpy().reshape(-1, 1))
    svr_rmse = np.sqrt(mean_squared_error(test_WRF_ts, svr_forecast))
    print(f'{kernel.capitalize()} SVR Model RMSE:', svr_rmse)
```
